## Ratios and Proportions

If two proportions are the same you can cross-multiply the two and should get the same result.

Example: 2/4 = 5/10 2 * 10 = 5 * 4 20

```python
# Enter 0 for the unknown value
n1 = 8
d1 = 8
n2 = 1
d2 = 0

if n2 == 0:
    answer = d2 * n1 / d1
    print( "n2 = ", answer)
else:
    answer = n2 * d1 / n1
    print("d2 = ", answer)
```

```python
# Get a decimal number and convert it
digits = input("Enter number to convert")

exponent = int(len(digits))-1
print("exp = ", exponent, sep="")

n = float(digits)
print("n = ", n, sep="")

numerator = int( n * 10**exponent)
print("numerator = ", numerator, sep="")

denominator = 10**exponent
print("denominator = ", denominator, sep="")

factor = 1
for i in range(1, denominator+1):
    if numerator % i == 0 and denominator % i == 0:
        factor = i

num = int(numerator / factor)
den = int(denominator / factor)

print("fraction = ", num, "/", den, sep="")

percent = n * 100
print("percent = ", percent, "%", sep="")
```

## Slope

Calculating the slope of a linear graph can be done if we know two points on  
``M = (y2 - y1)/(x2 - x1)``

```python
x1 = 1
x2 = 2
y1 = 1
y2 = 4

slope = (y2 - y1) / (x2 - x1)

print("Slope = ", slope)

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

ax.set_xlabel("x values")
ax.set_ylabel("y values")
ax.set_title("Graph slope")
ax.grid(True)
ax.set_xticks(np.arange(xmin, xmax, 1))
ax.set_yticks(np.arange(ymin, ymax, 1))

y = 3*x -2
plt.plot(x,y, "r",label='y=3x-2')
plt.legend()
plt.show()
```

Slope intercept formula:
``Y = mx + b``

```python
# Calculate a linear equation from two known points and plot it on a graph

x1 = 0
y1 = 2
x2 = 2
y2 = 4

# calculate slope:
m = (y2 - y1) / (x2 - x1)

# calculate slope intercept
b = y1 - m*x1

# Print equation:
print("y = ", m,"x+",b, sep="")

# graph data
xmin = -10
xmax = 10
ymin = -10
ymax = 10

# calculate start and stop points for the line
y_start = m*xmin + b
y_end = m*xmax + b

# basic graph canvas
fig, ax = plt.subplots()
plt.axis([xmin, xmax, ymin, ymax])
plt.plot([xmin, xmax], [0,0], "k")
plt.plot([0,0], [ymin, ymax], "k")
ax.grid(True)
ax.set_xticks(np.arange(xmin, xmax, 1))
ax.set_yticks(np.arange(ymin, ymax, 1))

# plot the linear function
plt.plot([xmin, xmax, ], [y_start, y_end], "b")
plt.show()
```

## Factoring

```python
print(5 % 3)
```

```python
num = 12

for i in range(1, num+1):
    if num % i == 0:
        print(i)
```

```python
numerator = 12
denominator = 26
factor = 1

# Find greatest common factor
for i in range(1, denominator+1):
    if numerator % i == 0 and denominator % i == 0:
        factor = i


# Divide out greatest common factor
n = int(numerator / factor)
d = int(denominator / factor)

print("Original: ", numerator, "/", denominator)
print("Reduced: ", n, "/", d)
```

```python
# Factoring out square roots

import math
import sympy
from sympy import symbols

# Number to factor
n = 12

max_factor = 1

upper_limit = math.floor(math.sqrt(n))+1

# find square factors
for i in range(1, upper_limit):
    if n % (i**2) == 0:
        max_factor = i

# Results
print("n = ", n, sep="")
print("Square rooted factor = ", max_factor, sep="")
print("Square factor = ", max_factor**2, sep="")
print("Integer = ", n/(max_factor**2), sep="")
```
