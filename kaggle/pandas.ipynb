# Pandas

Pandas is a widely used python library for data analysis.

```python
# Mostly used way to import pandas to python
import pandas as pd
```

## Core data objects in pandas

The main data objects in pandas are Dataframes and Series.

### DataFrames
DataFrames are tables consisting of an array of entries. Each entry corresponds to a row aand column:

```python
pd.DataFrame({"Yes": [10, 20], "No": [30, 40]})
```

```python
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 'Sue': ['Pretty good.', 'Bland.']})
```

```python
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 
              'Sue': ['Pretty good.', 'Bland.']},
             index=['Product A', 'Product B'])
```

### Series
Series are sequences of data. Compared to a DataFrame, a series would be an entity only containing one column.


```python
pd.Series([10, 20, 30, 40, 50])
```

```python
pd.Series([10, 20, 30, 40, 50], index = ["Sales 2015", "Sales 2016", "Sales 2017", "Sales 2018", "Sales 2019"], name = "Sales Product A")
```

### Indexing, loc and iloc

```python
# See the first eenrty obf a DataFrame
reviews.iloc[0]
```

```python
reviews.iloc[:, 0]
```

```python
reviews.iloc[2:5, :3]
```

```python
reviews.iloc[-5:, :3]
```

```python
reviews.loc[:, 'country']
```

```python
reviews.loc[:, ["country", "taster_name", "points"]]
```

```python
sample = [1, 2, 3, 5, 8]
sample_reviews = reviews.loc[sample]
sample_reviews
```

```python
sample = [0, 1, 10, 100]
reviews.loc[sample, ["country", "province", "region_1"]]
```

```python

reviews.loc[0:99, ["country", "variety"]]
```

### Manipulating the index

We can chnage the index with the set_index() method if we realise that we can use another index in the data:

```python
reviews.set_index("title")
```

### Conditional selection

```python
reviews.country == "Italy"
```

```python
reviews.loc[reviews.country == "Italy"]
```

```python
reviews.loc[(reviews.country == "Italy") & (reviews.points >= 90)]
```

```python
top_oceania_wines = reviews.loc[(reviews.country.isin(["Australia", "New Zealand"])) & (reviews.points >= 95)]
top_oceania_wines
```

```python
reviews.loc[(reviews.country == "Italy") | (reviews.points >= 90)]
```

```python
reviews.loc[(reviews.country.isin(["Italy", "France"])) & (reviews.points >= 90)]
```

```python
reviews.loc[reviews.price.notnull()]
```
