```python
import math
import sympy
from sympy import symbols, solve

def main ():
    menu()
    

def add():
  a = float(input("Enter a number: \n"))
  b = float(input("Enter the other number: \n"))
  sum = a + b
  print(a, "+", b, "=", sum)


def subtract():
  a = float(input("Enter a number: \n"))
  b = float(input("Enter the other number: \n"))
  sum = a - b
  print(a, "-", b, "=", sum)


def multiply():
  a = float(input("Enter a number: \n"))
  b = float(input("Enter the other number: \n"))
  sum = a * b
  print(a, "*", b, "=", sum)


def divide():
  a = float(input("Enter a number: \n"))
  b = float(input("Enter the other number: \n"))
  sum = a / b
  print(a, "/", b, "=", sum)


def prime():
  n = int(input("Enter an integer: "))
  p = "Prime"
  for i in range(2, n):
    if n % i == 0:
      p = "Not prime"
  print(p)


def prime_factors():
  n = int(input("Enter an integer: "))
  divisor = 2
  print(f"The prime factors of {n} are:")
  for divisor in range(2, int(n) + 1):
    if n % divisor == 0:
      print(divisor)
      n //= divisor


def simplify_sqrt():
  n = int(input("Enter an integer: "))
  upper_limit = math.floor(math.sqrt(n)) + 1
  max_factor = 1
  other_factor = 1
  square_root = 1

  for maybe_factor in range(1, upper_limit):
    if n % (maybe_factor**2) == 0:
        max_factor = maybe_factor**2

  other_factor = n/max_factor

  square_root = int(math.sqrt(max_factor))
  other_factor = int(other_factor)
  output = square_root*sympy.sqrt(other_factor)
  print(sympy.pretty(output))


def solve_for_x():
  eq = input('Enter an equation to solve for x: 0 = ')
  x = symbols('x')

  solution = solve(eq, x)
  for s in solution:
      print("x =", s)


def convert():
  digits = input("Enter a decimal number to convert: ")
  exponent = int(len(digits))-1
  n = float(digits)

  numerator = int(n * 10**exponent)
  denominator = 10 ** exponent
  percent = n * 100

  factor = 1
  for i in range(1, denominator+1):
    if numerator % i == 0 and denominator % i == 0:
      factor = i

  num_reduced = int(numerator / factor)
  den_reduced = int(denominator / factor)

  print("The fraction is ", num_reduced, "/", den_reduced, sep="")
  print("The percent is ", percent, "%", sep="")


def menu():
    while True:
        print("Menu:", "\n") # Header and a blank line
        print("1 Addition")
        print("2 Subtraction")
        print("3 Multiplication")
        print("4 Division")
        print("5 Check for prime")
        print("6 Find prime factors")
        print("7 Simplify a sqare root")
        print("8 Solve for x")
        print("9 Convert a decimal number into fractions and percent")
        print("q Quit")
        
        option = input("\nEnter option:\n")
        print("") # Empty line for niceness
    
        if option == "1":
            add()
        elif option == "2":
            subtract()
        elif option == "3":
            multiply()
        elif option == "4":
            divide()
        elif option == "5":
            prime()
        elif option == "6":
            prime_factors()
        elif option == "7":
            simplify_sqrt()
        elif option == "8":
            solve_for_x()
        elif option == "9":
            convert()
        elif option == "q":
            print("Goodbye!")
            break
        else:
            print("Not an option")


main()
```
